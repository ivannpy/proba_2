```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```python
def phi(u):
    """ Tranformación que convierte dos números generados pseudo-aleatoriamente
    con distribución uniforme (0,1) en dos números pseudo-aleatorios con 
    distribución normal estándar.
    
    Implementa la función
    f:R2 -> R2
    (x1, x2) --> (sqrt(-2 log(x1))cos(2 pi x2), sqrt(-2 log(x1))sin(2 pi x2))
    
    :param float u: 2-vector uniforme (0, 1)
    :return: Vector binormal estándar, (z1, z2).
    """
    aux = np.sqrt(-2 * np.log(u[0]))
    z1 = aux * np.cos(2 * np.pi * u[1])
    z2 = aux * np.sin(2 * np.pi * u[1])
    
    return z1, z2
```

```python
def generator(N):
    values_1 = []
    values_2 = []
    
    for i in range(N):
        # Generamos dos números con distribución uniforme (0,1)
        # U = (U1, U2)
        u = stats.uniform.rvs(0, 1, size=2)
        # Aplicamos la primera transformación
        # Z = (Z1, Z2)
        z = phi(u)
        values_1.append(z[0])
        values_2.append(z[1])
    
    return values_1, values_2
```

```python
N = 10_000
normal1, normal2 = generator(N)
```

```python
plt.figure(figsize=(10,7))
plt.hist(normal1, density=True, bins=300, color='blue')
plt.title('Densidad empírica de z1')
plt.savefig('hist.png')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.hist(normal2, density=True, bins=300, color='red')
plt.title('Densidad empírica de z1')
plt.savefig('hist.png')
plt.show()
```

```python
x_dom = np.linspace(-10, 10, 10000)
normal_pdf = stats.norm.pdf(x_dom)
```

```python
plt.figure(figsize=(10,7))
plt.plot(x_dom, normal_pdf, color='red', lw=2)
plt.hist(normal1, density=True, bins=100, color='blue')
plt.title('Densidad empírica de z1 y función de densidad')
plt.savefig('hist_density.png')
plt.show()
```
